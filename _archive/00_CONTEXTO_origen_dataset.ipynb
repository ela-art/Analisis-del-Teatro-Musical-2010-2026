# Origen del dataset maestro — Teatro Musical España (2010–2026)

Este documento describe el proceso de construcción del dataset `maestro_musicales.csv`, que actúa como punto de entrada del proyecto.

> **Este notebook no es ejecutable.** Documenta el proceso de recopilación y transformación de datos que se realizó manualmente. El pipeline ejecutable comienza en `01_ETL_star_schema.ipynb`.

## Descripción del dataset

`maestro_musicales.csv` es un dataset propio construido manualmente a partir de fuentes públicas del sector del teatro musical en España. Recoge **71 producciones estrenadas entre 2010 y 2026**, con las siguientes variables:

| Variable | Descripción |
|---|---|
| `obra` | Título del musical |
| `productora` | Empresa productora |
| `anio_inicio` | Año de estreno en España |
| `anio_fin` | Año de cierre (NaN si sigue en cartel) |
| `teatro` | Teatro principal de la producción |
| `gira` | Si realizó gira nacional (True/False) |
| `activa` | Si la producción sigue activa en 2026 (True/False) |
| `duracion` | Duración del espectáculo en minutos |
| `genero` | Género teatral (Drama, Comedia, Familiar, Biográfico) |
| `origen` | Franquicia internacional o Creación Propia |

## Fuentes utilizadas

- Webs oficiales de productoras: Stage Entertainment, Let's Go Company, ATG Entertainment, Beon Entertainment, Nostromo Live, entre otras
- Carteleras históricas: carteleramusicales.es
- Comunicación institucional y notas de prensa
- Conocimiento sectorial verificado manualmente

Los datos fuente originales (archivos Excel de trabajo) no se incluyen en el repositorio por contener información de elaboración propia no publicable.
