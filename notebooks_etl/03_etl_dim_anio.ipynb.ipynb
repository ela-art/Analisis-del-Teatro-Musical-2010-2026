```python
# librerías
import pandas as pd

# cargar tabla de hechos
fact = pd.read_csv("../data_processed/fact_producciones.csv")

# crear dimensión año (valores únicos)
dim_anio = fact[["anio_inicio"]].drop_duplicates().sort_values("anio_inicio")

# renombrar columna
dim_anio = dim_anio.rename(columns={"anio_inicio": "anio"})

# exportar
dim_anio.to_csv("../data_processed/dim_anio.csv", index=False, encoding="utf-8-sig")

# ver resultado
dim_anio.head()
```
